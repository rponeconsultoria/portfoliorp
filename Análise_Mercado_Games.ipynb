```python
# Recalculando 'Analise'

Analise = Base_Dados.groupby(by=['Year'])[['North America', 'Europe', 'Japan', 'Rest of World', 'Global']].sum()

#Analisando a propor√ß√£o dos 100% de cada continente comparado ao total

America =[America / Total * 100 for America, Total in zip(Analise['North America'], Analise['Global'] ) ]
Europa =[Europa / Total * 100 for Europa, Total in zip(Analise['Europe'], Analise['Global'] ) ]
Japao =[Japao / Total * 100 for Japao, Total in zip(Analise['Japan'], Analise['Global'] ) ]
Mundo =[Mundo / Total * 100 for Mundo, Total in zip(Analise['Rest of World'], Analise['Global'] ) ]

America, Europa, Japao, Mundo
```

```python
# Gr√°fico de % de Pa√≠ses em rela√ß√£o ao valor Global

#Tamanho
plt.figure( figsize=(10,5))

# Gr√°fico de % de Pa√≠ses em rela√ß√£o ao valor Global

#Tamanho
plt.figure(figsize=(10, 5))

#Largura da barra
Largura_Barra = 0.85

# Acessando o index do DataFrame
Rotulos = Analise.index
Grupos = ['America', 'Europa', 'Japao', 'Mundo']

# Criando um DataFrame para facilitar o plot
df_plot = pd.DataFrame({'America': America, 'Europa': Europa, 'Japao': Japao, 'Mundo': Mundo}, index=Rotulos)

# Plotando os dados
df_plot.plot(kind='bar', width=Largura_Barra)

# Titulo
plt.title('An√°lise de distribui√ß√£o por continente')
plt.xlabel('Ano')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation=0)  # Para evitar a rota√ß√£o dos r√≥tulos do eixo x

plt.show()
```

```python
#Observe que este gr√°fico √© dif√≠cil de se intepretar, vamos transformar valores quantitativos em qualitativos!

from sklearn.preprocessing import LabelEncoder

Funcao_Label = LabelEncoder()

Base_Dados['Produtor'] =Funcao_Label.fit_transform(Base_Dados['Publisher'])
Base_Dados['G√™nero'] =Funcao_Label.fit_transform(Base_Dados['Genre'])
Base_Dados['Jogo'] = Funcao_Label.fit_transform(Base_Dados['Game'])


Base_Dados.head()

#Esta fun√ß√£o pega todos os itens de Publishers e os transforma em numeros
```
