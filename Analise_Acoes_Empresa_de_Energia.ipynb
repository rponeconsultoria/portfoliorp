```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
Base_Dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Verificando
Base_Dados.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Gráfico

# Tamanho da figura
plt.figure(figsize=(15, 6))

# Título do gráfico
plt.title('Análise das ações de empresas de energia')

# Plota os dados da Petrobras
plt.plot(Base_Dados.index, Base_Dados['Petrobras'], color ='#008c4a', linewidth=4)

# Plota as demais colunas com transparência
for coluna in Base_Dados.columns[1:]:
    plt.plot(Base_Dados.index, Base_Dados[coluna], alpha=0.3, linewidth=3)

# Texto indicando Petrobras no final da linha
plt.text(Base_Dados.index[-1], Base_Dados['Petrobras'].tail(1).iloc[0], 'Petrobras',
         color ='#008c4a', size='large', horizontalalignment='left')

# Texto indicando Outros no final da linha
plt.text(Base_Dados.index[-1], Base_Dados[coluna].tail(1), coluna,
         color ='gray', size='large', horizontalalignment='left')

# Labels dos eixos
plt.xlabel('Periodo')
plt.ylabel('Preço do Fechamento (R$)')

# Mostrar o gráfico
plt.show()
```
