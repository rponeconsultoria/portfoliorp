```python
#Libs Essenciais

# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para análises
import matplotlib.pyplot as plt
import seaborn as sns

# Libs para ignorar avisos
import warnings

# Desabilitando os avisos
warnings.filterwarnings('ignore')
```

```python
#Lendo database
Base_Dados = pd.read_csv('Dados_Pib.xlsx - Tab_Dados.csv')

#Verificando
Base_Dados.head(10)
```

```python
# Agrupando e calculando a média apenas para colunas numéricas
Base_Dados.groupby(['Ano', 'PIB per capita']).mean(numeric_only=True)
```

```python
#Sistema de Grids

#Cor Fundo de Fundo
Cor_Fundo = '8f5f5f5'

#Criar o sistema de grids
Grid_Graficos= sns.FacetGrid( Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)
```

```python
#Adicionando gráfico de linhas em cada grafico
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

#Filtrar Título
Grid_Graficos = Grid_Graficos.set_titles(col_template='{col_name}')
plt.show()


#Adicionar Subtítulo
Grid_Graficos = Grid_Graficos.fig.suptitle('Evolução do Comparativo de renda per Capta x Estado')

#Ajustando
plt.subplots_adjust(top=0.9)
```

```python
# Configurar o estilo do gráfico
sns.set_theme(style="whitegrid")

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))  # Tamanho da figura
sns.barplot(
    x='Territorialidades',
    y='PIB per capita',
    data= Base_Dados,
    palette='viridis'
)

# Personalizar os eixos e título
plt.xlabel('Estados', fontsize=12)
plt.ylabel('Renda per capita (R$)', fontsize=12)
plt.title('Renda per capita por Estado', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotacionar os nomes dos estados para melhor visualização
plt.tight_layout()  # Ajustar o layout para evitar sobreposição

# Exibir o gráfico
plt.show()
```
