```python
#Libs Essenciais

# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para an√°lises
import matplotlib.pyplot as plt
import seaborn as sns

# Libs para ignorar avisos
import warnings

# Desabilitando os avisos
warnings.filterwarnings('ignore')


#Lendo o dataframe
Base_Dados = pd.read_csv('Salary_Data.csv')

#Verificando
Base_Dados.head()
```

```python
# Renomeando as colunas
Base_Dados.rename( columns={
    'YearsExperience': 'Anos de Experiencia',
    'Salary': 'Salario'
}, inplace=True)
```

```python
#Dimensao
Base_Dados.shape
```

```python
#Campos nulos
Base_Dados.isnull().sum()
```

```python
#Gr√°fico de Nulos
sns.heatmap(Base_Dados.isnull(), cbar=False)

#Obs: Podemos concluir que as linhas e colunas est√£o todas preenchidas
```

```python
Base_Dados.describe()

#Obs: podemos notar que temos desde 1 a 10 anos de experiencia na empresa, e a m√©dia salarial √© 37.000 U$D/Ano.
```

```python
#Gr√°fico de Renda

plt.figure( figsize=(10,8))
plt.title('An√°lise de Renda dos Funcion√°rios')
sns.kdeplot( Base_Dados['Salario'], color='red', linewidth=2, shade=True)
plt.show()
```

```python
#Gr√°fico de Experi√™ncia

plt.figure( figsize=(10,8))
plt.title('An√°lise de Experi√™ncia dos Funcion√°rios')
sns.kdeplot( Base_Dados['Anos de Experiencia'], color='blue', linewidth=2, shade=True)
plt.show()
```

```python
#An√°lise de Distribui√ß√£o - Sal√°rio

plt.figure( figsize=(10,4))
sns.boxplot( Base_Dados['Salario'])
```

```python
#An√°lise de Distribui√ß√£o - Experi√™ncia

plt.figure( figsize=(10,4))
sns.scatterplot( data=Base_Dados,x='Anos de Experiencia', y='Salario')

#Obs: Quanto mais tempo de experiencia, maior o sal√°rio do funcion√°riuo, √© o que podemos observar com este gr√°fico comparativo
```

```python
sns.regplot( data=Base_Dados,x='Anos de Experiencia', y='Salario')

#Obs: √â n√≠tido a rela√ß√£o entre as duas variaveis
```

```python
#Analise de Correla√ß√£o
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(5,5))
sns.heatmap( Correlacao, annot=True, cbar=False);
```
